```
data=[]
for i in os.listdir("/content/drive/My Drive/data"):
    try:
        df = pd.read_pickle('/content/drive/My Drive/data/'+i)
        print(i,", SIZE: {:,} bytes".format(sys.getsizeof(df)))
        data.append(df)
            
    except Exception as e:
        print(i, "Error loading file", repr(e))
         
        
```

```
data
```

```
import tensorflow as tf
```

```
action=[]
audio=[]
cast=[]
place=[]
for i in range(len(data)):
  feature_action_proto=tf.make_tensor_proto(data[i]['action'])
  action_feature=tf.make_ndarray(feature_action_proto)
  action_feature=action_feature.flatten()
  action.append(action_feature)
  feature_audio_proto=tf.make_tensor_proto(data[i]['audio'])
  audio_feature=tf.make_ndarray(feature_audio_proto)
  audio_feature=audio_feature.flatten()
  audio.append(audio_feature)
  feature_cast_proto=tf.make_tensor_proto(data[i]['cast'])
  cast_feature=tf.make_ndarray(feature_cast_proto)
  cast_feature=cast_feature.flatten()
  cast.append(cast_feature)
  feature_place_proto=tf.make_tensor_proto(data[i]['place'])
  place_feature=tf.make_ndarray(feature_place_proto)
  place_feature=place_feature.flatten()
  place.append(place_feature)

 
```

```
ground_truth=[]
predict=[]
shot_frame=[]
imdb_id=[]

for i in range(len(data)):
  ground_proto=tf.make_tensor_proto(data[i]['scene_transition_boundary_ground_truth'])
  ground=tf.make_ndarray(ground_proto)
  ground1=ground*1
  ground1=ground1.flatten()
  
  ground_truth.append(ground1)
  
  predict_proto=tf.make_tensor_proto(data[i]['scene_transition_boundary_prediction'])
  predict1=tf.make_ndarray(predict_proto)
  predict1=predict1.flatten()
  predict.append(predict1)

  shot_proto=tf.make_tensor_proto(data[i]['shot_end_frame'])
  shot=tf.make_ndarray(shot_proto)
  shot=shot.flatten()
  shot_frame.append(shot)

  imdb=data[i]['imdb_id']
  imdb_id.append(imdb)

  
```

```
Data=pd.DataFrame({'imdb_id':imdb_id,'action':action,'audio':audio,'cast':cast,'place':place,'scene_transition_boundary_ground_truth':ground_truth,'scene_transition_boundary_prediction':predict,'shot_end_frame':shot_frame})
```
